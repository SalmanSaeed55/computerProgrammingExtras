1. Create a program that asks for a user's name and their birth year. Output a message saying: "Hello [Name], you will turn [Age] this year." Ensure the age is an integer calculation.

```python
try:
    name = input("Please enter your name: ")
    birth_year = int(input("Please enter your birth year: "))
except  ValueError:
    print("Please enter a valid age")
else:
    age = 2026 - birth_year
    print(f"Hello {name}, you will turn {age} this year.")
```

2. Write a "Currency Converter" that takes a float input (amount in GBP) and converts it to USD (use a fixed rate like 1.27). Format the output to exactly 2 decimal places using an f-string.

```python
GBP_TO_USD = 1.27

try:
    gbp = float(input("Enter the amount in GBP: "))
except ValueError:
    print("Please enter a valid amount.")
else:
    usd = gbp * GBP_TO_USD
    print(f"${gbp:.2f} is equal to £{usd:.2f} USD.")
```

1. Create a list of 10 random numbers. Write a for loop that iterates through the list and prints "High" if the
number is above 50, "Low" if it is below 50, and "Perfect" if it is exactly 50.

```python
numbers = [23, 67, 45, 50, 89, 12, 34, 56, 78, 90]

for num in numbers:
    if num > 50:
        print(f"{num} is High")
    elif num < 50:
        print(f"{num} is Low")
    else:
        print(f"{num} is Perfect")
```

2. Use a while loop to create a simple login system. Ask the user for a password. If they type "Python123", print "Access Granted" and break the loop. Otherwise, keep asking.

```python
password = "Python123"

while True:
    user_password = input("Please enter the password: ")

    if user_password == password:
        print("Access granted")
        break
    else:
        print("Incorrect password, please try again.")
```

1. Write a function called calculate_bmi(weight, height). It should return the BMI value ($BMI = weight / height^2$).

```python
def calculate_bmi(weight, height):
    try:
        bmi = weight / (height ** 2)
        return f"{bmi:.2f} BMI"
    except ZeroDivisionError:
        return "Height cannot be zero."
    except TypeError:
        return "Please enter valid numbers for weight and height."

print(calculate_bmi(70, 1.75))
```

2. Wrap an input() call in a try/except block. If the user enters a string instead of a number for a calculation, catch the ValueError and print a friendly warning instead of letting the program crash.

```python
try:
    number = int(input("Please enter a number: "))
    result = number * 2
    print(f"The result of doubling your number is: {result}")
except ValueError:
    print("That's not a valid number. Please enter a numeric value.")
```

1. Create a dictionary representing a "Shopping Cart" where the keys are item names and the values are prices. Write a loop that calculates the total price of all items in the dictionary.

```python
shopping_cart = {
    "Apples": 3.50,
    "Bread": 2.00,
    "Milk": 1.50,
    "Eggs": 2.75
}
total_price = 0

for item, price in shopping_cart.items():
    total_price += price

print(f"Total price: ${total_price:.2f}")
```

2. Simulate file handling. Write a script that "reads" a list of names (you can just create a list variable) and saves only the names starting with the letter "A" into a new list called filtered_names. (If you have a .txt file handy, try to actually use with open()).

```python
names = ["Alice", "Bob", "Amanda", "Charlie", "Andrew", "David"]
filtered_names = []

for name in names:
    if name.startswith("A"):
        filtered_names.append(name)

print(filtered_names)
```
